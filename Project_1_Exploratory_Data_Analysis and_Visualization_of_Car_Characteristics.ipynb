# Exploratory Data Analysis and Visualization of Car Characteristics

## Project Introduction:

This project aims to explore and analyze a comprehensive dataset of car characteristics, encompassing a wide range of features that contribute to the overall performance, efficiency, and pricing of various car models. By conducting a thorough exploratory data analysis (EDA) and creating informative visualizations, the goal is to uncover valuable insights that can inform decision-making for car buyers, manufacturers, and industry stakeholders.

### Fiche Descriptive of Dataframe Columns:

**symboling**: This column represents the insurance risk rating of the car. It is an integer value ranging from -3 to 3, where a higher positive value indicates a higher risk rating and a lower negative value indicates a lower risk rating.

**normalized-losses**: This column represents the percentage of the car's value lost due to depreciation. It is a numeric value that reflects the car's relative loss in value over time.

**make**: This column specifies the manufacturer or brand of the car.

**fuel-type**: This column indicates the type of fuel used by the car, such as gas or diesel.

**aspiration**: This column describes the engine's air intake system, which can be either turbo-charged or naturally aspirated.

**num-of-doors**: This column specifies the number of doors the car has.

**body-style**: This column describes the car's body style, such as sedan, hatchback, or convertible.

**drive-wheels**: This column indicates the type of drive wheels, such as front-wheel drive, rear-wheel drive, or four-wheel drive.

**engine-location**: This column specifies the location of the engine within the car, either front or rear.

**wheel-base**: This column represents the distance between the front and rear wheels of the car, measured in inches.

**length**: This column specifies the overall length of the car, measured in inches.

**width**: This column represents the overall width of the car, measured in inches.

**height**: This column indicates the overall height of the car, measured in inches.

**curb-weight**: This column represents the weight of the car, including all its standard equipment and a full tank of fuel, measured in pounds.

**engine-type**: This column describes the type of engine used in the car, such as overhead cam or rotary.

**num-of-cylinders**: This column specifies the number of cylinders in the car's engine.

**engine-size**: This column represents the displacement of the car's engine, measured in cubic centimeters (cc).

**fuel-system**: This column indicates the type of fuel system used in the car, such as electronic fuel injection or carburetor.

**bore**: This column represents the diameter of the engine's cylinders, measured in millimeters.

**stroke**: This column specifies the distance the piston travels within the engine's cylinders, measured in millimeters.

**compression-ratio**: This column represents the ratio of the volume of the engine cylinder from its maximum volume to its minimum volume.

**horsepower**: This column indicates the maximum power output of the car's engine, measured in horsepower.

**peak-rpm**: This column specifies the engine speed at which the maximum horsepower is achieved, measured in revolutions per minute (RPM).

**city-mpg**: This column represents the car's fuel efficiency in the city, measured in miles per gallon (MPG).

**highway-mpg**: This column indicates the car's fuel efficiency on the highway, measured in miles per gallon (MPG).

**price**: This column specifies the price of the car, which is the target variable in this dataset.

By exploring and analyzing this comprehensive dataset, the project aims to uncover valuable insights and patterns that can help stakeholders make informed decisions about car characteristics, performance, and pricing.

## Data Wrangling 

Handling missing values is a critical aspect of data preprocessing in data analysis and machine learning. When working with a DataFrame, missing values can occur due to various reasons such as incomplete data collection, data corruption, or data processing errors. Identifying and dealing with these missing values is essential to ensure the accuracy and reliability of analytical results. Pandas, a popular data manipulation library in Python, provides several methods to detect and handle missing data. The isnull() method allows you to identify which values are missing, returning a DataFrame of Boolean values indicating the presence of NaN (Not a Number) or None values. Conversely, the notnull() method can be used to identify non-missing values. Depending on the analysis and the dataset, common strategies for handling missing values include imputation (replacing missing values with a statistical estimate like mean, median, or mode), deletion of rows or columns with missing data, or using advanced techniques such as interpolation based on neighboring values. Effective management of missing data ensures that analytical models are robust and accurately reflect underlying patterns within the data.

```python
print('The number of missing values for each column:')
df.isna().sum()
```

The isna() method applied to df returns a boolean DataFrame where True indicates a missing value (NaN or None) and False indicates a non-missing value. By chaining .sum() after isna(), Pandas calculates the number of True values (missing values) in each column across the entire DataFrame. This provides a quick summary of missing data per column, which is essential for data quality assessment and preprocessing. Understanding the distribution and extent of missing values helps in making informed decisions on how to handle them, whether through imputation techniques, deletion of affected rows or columns, or other data cleaning strategies. This approach is particularly useful in exploratory data analysis to identify potential data gaps and ensure the robustness of subsequent analytical or modeling tasks.

Now, we need to replace the "?" symbol with NaN so the dropna() can remove the missing values:


```python
# Replace all '?' values with NaN 
df.replace("?",np.nan,inplace = True)
```

We should only drop whole columns if most entries in the column are empty. In the data set, none of the columns are empty enough to drop entirely.

<b>Replace by mean:</b>
<ul>
    <li>"normalized-losses": 41 missing data, we will replace them with mean</li>
    <li>"stroke": 4 missing data, we will replace them with mean</li>
    <li>"bore": 4 missing data, we will replace them with mean</li>
    <li>"horsepower": 2 missing data, we will replace them with mean</li>
    <li>"peak-rpm": 2 missing data, we will replace them with mean</li>
</ul>

```python
# Calculate the average of 'normalized-losses', 'stroke', 'bore', 'horsepower', and 'peak-rpm' columns

average_normalized_losses = df["normalized-losses"].astype(float).mean(axis=0)
average_stroke = df["stroke"].astype(float).mean(axis=0)
average_bore = df["bore"].astype(float).mean(axis=0)
average_horsepower = df["horsepower"].astype(float).mean(axis=0)
average_peak_rpm = df["peak-rpm"].astype(float).mean(axis=0)

# Output the calculated averages

print("Average of normalized-losses :", average_normalized_losses)
print("Average of stroke :", average_stroke)
print("Average of bore :", average_bore)
print("Average of horsepower :", average_horsepower)
print("Average of peak-rpm :", average_peak_rpm)
```

```python
# Replace missing values (NaN) in specific columns with their respective calculated averages

df["normalized-losses"].replace(np.nan, average_normalized_losses, inplace=True)
df["stroke"].replace(np.nan, average_stroke, inplace=True)
df["bore"].replace(np.nan, average_bore, inplace=True)
df["horsepower"].replace(np.nan, average_horsepower, inplace=True)
df["peak-rpm"].replace(np.nan, average_peak_rpm, inplace=True)
```

```python
# Check
# Display the number of missing values for each column in the DataFrame
print('The number of missing values for each column:')
df.isna().sum()
```

We observe that the missing values have been removed from the following columns :

+ peak-rpm

## Correct data format
We are almost there!
<p>The last step in data cleaning is checking and making sure that all data is in the correct format (int, float, text or other).</p>

In Pandas, we use:
<p><b>.dtype()</b> to check the data type</p>
<p><b>.astype()</b> to change the data type</p>

Let's list the data types for each column

```python
df.dtypes
```

<p>As we can see above, some columns are not of the correct data type. Numerical variables should have type 'float' or 'int', and variables with strings such as categories should have type 'object'. For example, the numerical values 'bore' and 'stroke' describe the engines, so we should expect them to be of the type 'float' or 'int'; however, they are shown as type 'object'. You have to convert data types into a proper format for each column using the "astype()" method.</p> 

```python
# Convert data types to proper format

df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

## Visualizations 

### Comprehensive Analysis of Car Characteristics using Pie Charts

```python
# Create evenly spaced bins between the minimum and maximum values of the 'price' column

binn = np.linspace(min(df['price']), max(df['price']), 4)

# Define the labels for the bins

groupes = ['Low', 'Median', 'High']

# Create a new column 'binned_price' in the DataFrame, 

df["binned_price"] = pd.cut(df['price'], bins = binn, labels = groupes)
```

```python
# Calculate the count of each car body style and create a dataframe

style = df['body-style'].value_counts().reset_index()


# Calculate the count of each number of doors and create a dataframe 

doors = df['num-of-doors'].value_counts().reset_index()


# Calculate the count of each fuel type and create a dataframe

fuel = df['fuel-type'].value_counts().reset_index()


# Calculate the count of each drive wheel type and create a dataframe

wheels = df['drive-wheels'].value_counts().reset_index()


# Calculate the count of each engine type and create a dataframe

engine = df['engine-type'].value_counts().reset_index()

# Calculate the count of each binned price range and create a dataframe

price = df["binned_price"].value_counts().reset_index()
```

```python
# Create a figure

fig = plt.figure(figsize=(14, 10))

# Add six subplots arranged in a 2x3 grid

ax1 = fig.add_subplot(2, 3, 1)
ax2 = fig.add_subplot(2, 3, 2)
ax3 = fig.add_subplot(2, 3, 3)
ax4 = fig.add_subplot(2, 3, 4)
ax5 = fig.add_subplot(2, 3, 5)
ax6 = fig.add_subplot(2, 3, 6)

# Explode lists to highlight specific slices in the pie charts

explode_list1 = [0, 0, 0, 0, 0.2]
explode_list2 = [0, 0.1]
explode_list3 = [0, 0.1]
explode_list4 = [0, 0, 0.1]
explode_list5 = [0, 0, 0, 0, 0, 0.1]
explode_list6 = [0, 0.1, 0.2]

# Color lists to customize the appearance of the pie charts

color_list1 = ['gold', 'greenyellow', 'skyblue', 'pink', 'salmon']
color_list2 = ['skyblue', 'salmon']
color_list3 = ['pink', 'skyblue']
color_list4 = ['aqua' ,'salmon', 'gold']
color_list5 = ['skyblue', 'gold' , 'pink','salmon', 'aqua', 'greenyellow']
color_list6 = ['salmon', 'greenyellow', 'aqua']

# Plot 1: Pie Chart of Car Body Styles

ax1.pie(style['count'], autopct = '%1.1f%%', startangle = 90, explode = explode_list1, colors = color_list1)
ax1.axis('equal')
ax1.set_title('Pie Chart of Car Body Styles')
ax1.legend(fontsize = 6, loc = "upper left", labels = style["body-style"])

# Plot 2: Pie Chart of Car Doors

ax2.pie(doors['count'], autopct = '%1.1f%%', startangle = 90, explode = explode_list2, colors = color_list2)
ax2.axis('equal')
ax2.set_title('Pie Chart of Car Doors')
ax2.legend(fontsize = 8, loc = "upper left", labels = doors['num-of-doors'])

# Plot 3: Pie Chart of Fuel Type

ax3.pie(fuel['count'], autopct = '%1.1f%%', startangle = 90, explode = explode_list3, colors = color_list3)
ax3.axis('equal')
ax3.set_title('Pie Chart of Fuel Type')
ax3.legend(fontsize = 8, loc = "upper left", labels = fuel['fuel-type'])

# Plot 4: Pie Chart of Car Wheels

ax4.pie(wheels['count'], autopct = '%1.1f%%', startangle = 90, explode = explode_list4, colors = color_list4)
ax4.axis('equal')
ax4.set_title('Pie Chart of Car Wheels')
ax4.legend(fontsize = 8, loc = "upper left", labels = wheels['drive-wheels'])

# Plot 5: Pie Chart of Engine Type

ax5.pie(engine['count'], autopct = '%1.1f%%', startangle = 90, explode = explode_list5, colors = color_list5)
ax5.axis('equal')
ax5.set_title('Pie Chart of Engine Type')
ax5.legend(fontsize = 8, loc = "upper left", labels = engine['engine-type'])

# Plot 6: Pie Chart of Car Price

ax6.pie(price['count'], autopct = '%1.1f%%', startangle = 90, colors = color_list6, explode = explode_list6)
ax6.axis('equal')
ax6.set_title('Pie Chart of Car Price')
ax6.legend(fontsize = 8, loc = "upper left", labels = price["binned_price"])

# Display the figure

plt.show()
```

### Visual Analysis of Body Style, Number of Doors, and Their Relationship in Automobile Data

```python
# Create a figure

fig = plt.figure(figsize=(14, 6))

# Add three subplots arranged in a 1x3 grid

ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)

# Plot 1: Count plot for 'body-style'

sns.countplot(data = df, x = 'body-style', ax = ax1)
ax1.set_xlabel('Body Style')
ax1.tick_params(axis = 'x', rotation = 90)
ax1.set_ylabel('Frequency')
ax1.set_title('Frequency of Body Styles')

# Plot 2: Count plot for 'num-of-doors'

sns.countplot(data = df, x = 'num-of-doors', ax = ax2)
ax2.set_xlabel('Number of Doors')
ax2.set_ylabel('Frequency')
ax2.set_title('Frequency of Number of Doors')

# Plot 3: Count plot for 'body-style' with 'num-of-doors' as hue

sns.countplot(data = df, x = 'body-style', hue = 'num-of-doors', ax = ax3)
ax3.set_xlabel('Body Style')
ax3.tick_params(axis = 'x', rotation = 90)
ax3.set_ylabel('Frequency')
ax3.set_title('Body Style Frequency by Number of Doors')

# Adjust layout to prevent overlapping of subplots

plt.tight_layout()

# Display the plots

plt.show()
```

The code utilizes Matplotlib and Seaborn to visualize the frequency distributions of automotive data stored in the DataFrame df. It creates a figure with three subplots arranged horizontally. The first subplot displays a count of different 'body-style' categories among the vehicles in the dataset, providing insights into the distribution of car body types such as sedan, hatchback, and convertible. The second subplot presents the frequency of cars based on the number of doors they have, shedding light on the popularity of two-door versus four-door configurations. The third subplot enhances the analysis by plotting 'body-style' categories again but this time with 'num-of-doors' as a distinguishing factor using different colors, offering a comparative view of door counts across different body styles. Together, these visualizations enable viewers to quickly grasp the distribution patterns within categorical variables related to automobile characteristics, facilitating informed insights for further analysis or decision-making in automotive research or market analysis contexts.

### Car Price Distribution and Categorization

```python
# Create bins for the 'price' column with equally spaced intervals from min to max price

binn = np.linspace(min(df['price']), max(df['price']), 4)

# Define labels for the price bins

groups = ['Low', 'Medium', 'High']

# Create a new column 'binned_price' by categorizing 'price' into the defined bins with labels

df['binned_price'] = pd.cut(x = df["price"], bins = binn, labels = groups)
```

```python
# Create a figure

fig = plt.figure(figsize = (14, 6))

# Add two subplots arranged in a 1x2 grid

ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

# Plot 1: Histogram for 'price' with a kernel density estimate (KDE)

sns.histplot(df['price'], bins = 30, kde = True, ax = ax1)
ax1.set_xlabel('Price')
ax1.set_ylabel('Count')
ax1.set_title('Distribution of Car Prices')

# Plot 2: Count plot for 'binned_price' to show the count of cars in each price category

sns.countplot(data = df, x = 'binned_price', ax = ax2)
ax2.set_xlabel('Binned Price Categories')
ax2.set_ylabel('Count')
ax2.set_title('Count of Cars by Binned Price Categories')


# Adjust layout to prevent overlapping of subplots

plt.tight_layout()

# Display the plots

plt.show()
```

The results of the code provide a comprehensive analysis of car prices through two visualizations. The first subplot displays a histogram of car prices, overlaid with a kernel density estimate (KDE) curve, offering insights into the overall distribution and density of prices. This visualization helps identify the range, central tendency, and any skewness or kurtosis in the data. The second subplot categorizes car prices into three equally spaced intervals: Low, Medium, and High. It then presents a count plot showing the number of cars in each price category. This categorization and count plot provide a clearer understanding of the distribution of cars across different price ranges, highlighting the relative abundance or scarcity of cars within each category. Together, these visualizations offer valuable insights into the structure and distribution of car prices in the dataset.

### Comparative Analysis of Car Prices by Brand, Fuel Type, Drive Wheels, and Engine Type

```python
# Calculer le prix moyen par type de moteur

# Grouping the DataFrame 'df' by 'make' to calculate the mean price for each car brand

df_brand_price = df.groupby("make")[['price']].mean().reset_index()

# Grouping the DataFrame 'df' by 'engine-type' to calculate the mean price for each engine type

avg_price_by_engine_type = df.groupby('engine-type')['price'].mean().reset_index()
```

### Exploring Car Characteristics and Performance by Body Style and Cylinder Configuration

```python
# Group the dataframe by 'body-style' and calculate the mean 'curb-weight' for each group
# Reset the index to convert the resulting Series into a DataFrame

avg_curb_weight_by_body_style = df.groupby('body-style')['curb-weight'].mean().reset_index()

# Group the dataframe by 'num-of-cylinders' and calculate the mean 'horsepower' for each group
# Reset the index to convert the resulting Series into a DataFrame

df['horsepower'] = df['horsepower'].astype('float')
avg_horsepower_by_cylinders = df.groupby('num-of-cylinders')['horsepower'].mean().reset_index()
```

```python
# Create a figure

fig = plt.figure(figsize = (14, 6))

# Add four subplots arranged in a 2x2 grid

ax1 = fig.add_subplot(2, 2, 1)  # Subplot 1 for average curb weight by body style
ax2 = fig.add_subplot(2, 2, 2)  # Subplot 2 for car price distribution by number of doors and body style
ax3 = fig.add_subplot(2, 2, 3)  # Subplot 3 for average horsepower by number of cylinders
ax4 = fig.add_subplot(2, 2, 4)  # Subplot 4 for horsepower distribution by number of cylinders

# Plot 1: Bar plot for average curb weight by body style

sns.barplot(data = avg_curb_weight_by_body_style, x = 'body-style', y = 'curb-weight', ax = ax1)
ax1.set_xlabel('Body Style')
ax1.set_ylabel('Average Curb Weight')
ax1.set_title('Average Curb Weight by Body Style')

# Plot 2: Box plot for car price distribution by number of doors and body style
sns.boxplot(data = df, x = 'num-of-doors', y = 'price', hue = 'body-style', ax = ax2)
ax2.set_xlabel('Number of Doors')
ax2.set_ylabel('Price')
ax2.set_title('Car Price Distribution by Number of Doors and Body Style')
ax2.legend(fontsize = 7)

# Plot 3: Bar plot for average horsepower by number of cylinders
sns.barplot(data = avg_horsepower_by_cylinders, x = 'num-of-cylinders', y = 'horsepower', ax = ax3)
ax3.set_xlabel('Number of Cylinders')
ax3.set_ylabel('Average Horsepower')
ax3.set_title('Average Horsepower by Number of Cylinders')

# Plot 4: Box plot for horsepower distribution by number of cylinders
sns.boxplot(data = df, x = 'num-of-cylinders', y = 'horsepower', ax = ax4)
ax4.set_xlabel('Number of Cylinders')
ax4.set_ylabel('Horsepower')
ax4.set_title('Horsepower Distribution by Number of Cylinders')

# Adjust layout to prevent overlapping of subplots
plt.tight_layout()

# Display the plots
plt.show()
```
